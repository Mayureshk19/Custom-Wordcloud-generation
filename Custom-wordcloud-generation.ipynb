```python
wc = WordCloud()
```

```python
wc.generate(names)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(max_font_size=40).generate(names)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("1.jpg"))
wc = WordCloud(background_color="white", mask=custom_mask)
wc.generate(names)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("1.jpg"))
wc = WordCloud(background_color="white", mask=custom_mask, contour_width=1, contour_color='steelblue')
wc.generate(names)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
wc.to_file('mahadeva.jpg')
```

```python
custom_mask = np.array(Image.open("2.jpg"))
wc = WordCloud(background_color="white", mask=custom_mask, contour_width=1, contour_color='steelblue')
wc.generate(names)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func=image_colors)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
wc.to_file('mahakaal.jpg')
```
